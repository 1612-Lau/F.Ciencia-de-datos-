```python
import pandas as pd
import numpy as np

# Generar datos aleatorios
np.random.seed(42)
data = {
    "Sexo": np.random.choice(["M", "F"], size=30),
    "Edad": np.random.randint(18, 30, size=30),
    "Estatura": np.round(np.random.uniform(1.50, 1.80, size=30), 2),
    "Nota": np.round(np.random.uniform(1.0, 5.0, size=30), 2),
    "Ciudad": np.random.choice(["Armenia", "Barranquilla", "Cali", "Medellin"], size=30)
}

# Crear DataFrame
df = pd.DataFrame(data)
df.head(30)
```

```python
#1. Realizar una tabla de frecuencias absolutas y otra de frecuencias relativas para la
#variable Calificación. Almacena las tablas anteriores en dos variables y llámalas
#absolutas y relativas.

# Tabla de frecuencias absolutas
absolutas = df["Nota"].value_counts().sort_index()
print("Tabla de frecuencias absolutas:\n", absolutas)

# Tabla de frecuencias relativas
relativas = df["Nota"].value_counts(normalize=True).sort_index()
print("\nTabla de frecuencias relativas:\n", relativas)
```

```python
#2.Representar la variable ciudad mediante un diagrama de barras y un diagrama de
#sectores. Incluye un título adecuado para cada gráfico y colorea las barras y los
#sectores de colores diferentes.

import matplotlib.pyplot as plt

# Diagrama de barras
plt.figure(figsize=(8, 6))
df["Ciudad"].value_counts().plot(kind="bar", color=["blue", "pink", "red", "purple"])
plt.title("Distribución de estudiantes por Ciudad")
plt.xlabel("Ciudad")
plt.ylabel("Frecuencia")
plt.show()

# Diagrama circular
plt.figure(figsize=(8, 6))
df["Ciudad"].value_counts().plot(kind="pie", colors=["blue", "pink", "red", "purple"], autopct="%1.1f%%")
plt.title("Distribución de estudiantes por Ciudad")
plt.ylabel("")  # Ocultar etiqueta del eje Y
plt.show()
```

```python
#Para la variable Edad, realizar un histograma y un diagrama de caja y bigotes
#considerando la opción range = 1.5. Incluye un título apropiado para cada gráfico y
#colorea las barras del histograma de color amarillo. ¿Existe algún valor atípico en
#esta variable? Reduce el valor del argumento range hasta 0.5. ¿Varían las
#conclusiones?

# Histograma
plt.figure(figsize=(8, 6))
plt.hist(df["Edad"], bins=10, color="yellow")
plt.title("Histograma Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

# Diagrama de caja y bigotes (range=1.5)
plt.figure(figsize=(8, 6))
plt.boxplot(df["Edad"], vert=False, widths=0.7, flierprops=dict(markerfacecolor='r', marker='o'))
plt.title("Diagrama de caja y bigotes de Edad (range=1.5)")
plt.xlabel("Edad")
plt.show()

# Revisar si hay valores atípicos reduciendo el valor de range a 0.5
plt.figure(figsize=(8, 6))
plt.boxplot(df["Edad"], vert=False, widths=0.7, whis=0.5, flierprops=dict(markerfacecolor='r', marker='o'))
plt.title("Diagrama de caja y bigotes de Edad (range=0.5)")
plt.xlabel("Edad")
plt.show()
```

```python
#Realizar un resumen de la variable Puntuación mediante la orden summary.
#Comprueba que las medidas que proporciona summary coinciden con las medidas
#calculadas de forma individual usando su función específica

# Utilizar summary
resumen = df["Nota"].describe()
print("Resumen de la variable Nota:\n", resumen)

# Cálculo de medidas específicas
media_nota = df["Nota"].mean()
mediana_nota = df["Nota"].median()
min_nota = df["Nota"].min()
max_nota = df["Nota"].max()
varianza_nota = df["Nota"].var()

print("\nMedidas calculadas individualmente:")
print(f"Media: {media_nota}, Mediana: {mediana_nota}, Mínimo: {min_nota}, Máximo: {max_nota}, Varianza: {varianza_nota}")
```

```python
#Calcular la estatura media de los estudiantes y proporcionar al menos, dos
#medidas que indiquen la dispersión de esta variable.

# Calcular la estatura media
media_estatura = df["Estatura"].mean()
print("Estatura media:", media_estatura)

# Calcular medidas de dispersión
desviacion_estandar = df["Estatura"].std()
rango_estatura = df["Estatura"].max() - df["Estatura"].min()
varianza_estatura = df["Estatura"].var()

print(f"Desviación estándar: {desviacion_estandar}")
print(f"Rango: {rango_estatura}")
print(f"Varianza: {varianza_estatura}")
```
